# I.   Classes, Objets et Méthodes de Classe
Python est un langage qui permet la Programmation Orientée Objet (POO).

Une classe correspond à un modèle de données. Tous les objets d’une classe partagent les mêmes attributs (caractéristiques) qui définissent son état et les mêmes méthodes (actions appliquables aux instances. Elles prennent la forme de fonctions). Une classe peut être vue comme une famille d’objets.<br>

Un objet, ou instance de classe, est une entité particulière de cette classe définie par son état.<br>

**Exemple**: On peut créer une classe "Equation_Second_Degre" avec :<br>
- Comme attributs les coefficients a,b et c ($\ a x^2+b x+c = 0 $)
- Comme méthodes :
    - Le calcul du determinent
    - Le calcul des solutions réelles s'il y en a

```python
class Equation_Second_Degre :     # Création de la classe (class Nom_classe:) 
    def __init__(self, A, B,C):   # __init__ est une méthode particulière qu'on appelle un constructeur.
                                  # elle permet l'instanciation d'un objet avec les attributs fournis en arguments
                                  # l'argument self du constructeur fait référence à une instance générique de la classe
        self.a=A                  # initialisation des arguments
        self.b=B
        self.c=C
        
    def det(self):                # Méthode de classe permettant de claculer le determinent.
                                  # Il suffit de mettre self en argument car les coefficients 
                                  # a, b et c sont désormais des attributs associés à l'objet 
                                  # générique self
                    
        det=self.b*self.b-4*self.a*self.c
        return(det)
    
    def sol(self):                # Méthode de classe permettant de calculer les solutions réelles
        
        if self.det()==0:         # La méthode det est appliquée à l'objet générique self pour tester
                                  # la valeur du determinent
            sol=-self.b/(2*self.a)
            return sol
       
        elif self.det()>0:
            sol_1=(-self.b-self.det()**(1/2))/(2*self.a)
            sol_2=(-self.b+self.det()**(1/2))/(2*self.a)
            return sol_1,sol_2
        
        else:
            return None
```

```python
# Pour créer un instance de classe (un objet), il suffit de fournir des valeurs d'attribut à la classe
# Eq_1 est une instance de la classe Equation_Second_Degre avec comme attributs : 1, 4 et 1
Eq_1=Equation_Second_Degre(1,4,1)
```

```python
# Appliquer une méthode à une instance se fait simplement à travers une instruction de type :
# instance.méthode()
Eq_1.det()
```

```python
Eq_1.sol()
```

```python
# Si vous affichez la valeur d'une instance, Python renvoie son type et sont adresse mémoire
Eq_1
```

**Exercice 2**:<br>
Créez une classe de points géométriques dans l'espace euclidien à deux dimensions dotée d'une méthode permettant de calculer la distance entre une instance de cette classe et un autre point donné en argument.<br>
*Rappel:* la distance euclidienne entre deux points a et b dans un espace à deux dimensions s'obtient comme suit :<br>
<center>$\ \sqrt{(x_{a}-x_{b})^2+ (y_{a}-y_{b})^2} $</center> <br>

Créez deux instances p_1 et p_2 avec les coordonnées de votre choix<br>
Calculez la distance entre p_1 et p_2 en utilisant la méthode de classe.



```python
# Répondez en complétant de code ci-dessous
class Point:
    def __init__(self,Ab,Or): # ajouter des arguments si nécessaire
        self.abs=Ab
        self.ord=Or

    def dist(self,p): # ajouter des arguments si nécessaire
        dist=((self.abs-p.abs)**2+(self.ord-p.ord)**2)**1/2
        return(dist)
```

```python
# Instanciation de p_1 et p_2
p_1=Point(3,2)
p_2=Point(5,6)
```

```python
# Calcul de la distance
p_1.dist(p_2)
```

Un Dataframe est un objet qui peut être vu comme une collection de plusieurs objets élémentaires appelés Series.<br>
Une Serie n'est autre qu'un tableau à une seule dimension contenant des données indexées.<br>

Numpy est une librairie dédiée au calcul vectoriel, matriciel et tensoriel (tableaux multi-dimensionnel).<br>
Les structures de données manipulées par Numpy sont des numpy array.<br>
Contrairement aux listes qui pointent sur autant d'objet qu'elles contiennent d'éléments, le numpy array pointe sur un seul objet caractérisé entre autres par son contenu (data) et ses dimensions.<br>
